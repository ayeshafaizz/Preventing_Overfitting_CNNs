# Preventing Overfitting in CNNs

## Objective

Train and compare Baseline, Dropout, BatchNorm, and L2 CNN models on CIFAR-10.

## Import Libraries

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models, regularizers
```

## Load and Explore CIFAR-10 Dataset

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.cifar10.load_data()

print("Train:", X_train.shape)
print("Test :", X_test.shape)
```

## Data Preprocessing

```python
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

X_val = X_train[-5000:]
y_val = y_train[-5000:]

X_train_small = X_train[:-5000]
y_train_small = y_train[:-5000]

print(X_train_small.shape)
print(X_val.shape)
```

## Building the Baseline CNN

```python
def build_baseline():
    model = models.Sequential([
        layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
        layers.MaxPooling2D(),
        layers.Conv2D(64, (3,3), activation='relu'),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])
    return model
```

## Training the Baseline Model

```python
baseline_model = build_baseline()

baseline_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history_baseline = baseline_model.fit(
    X_train_small,
    y_train_small,
    validation_data=(X_val, y_val),
    epochs=15,
    batch_size=64
)
```

## CNN with Dropout

```python
def build_dropout():
    model = models.Sequential([
        layers.Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)),
        layers.MaxPooling2D(),
        layers.Conv2D(64,(3,3),activation='relu'),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(128,activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(10,activation='softmax')
    ])
    return model

dropout_model = build_dropout()
dropout_model.compile(optimizer='adam',
                      loss='sparse_categorical_crossentropy',
                      metrics=['accuracy'])

history_dropout = dropout_model.fit(
    X_train_small,
    y_train_small,
    validation_data=(X_val, y_val),
    epochs=15,
    batch_size=64
)
```

## CNN with Batch Normalization

```python
def build_bn():
    model = models.Sequential([
        layers.Conv2D(32,(3,3),input_shape=(32,32,3)),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D(),
        layers.Conv2D(64,(3,3)),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(128),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.Dense(10,activation='softmax')
    ])
    return model

bn_model = build_bn()
bn_model.compile(optimizer='adam',
                 loss='sparse_categorical_crossentropy',
                 metrics=['accuracy'])

history_bn = bn_model.fit(
    X_train_small,
    y_train_small,
    validation_data=(X_val, y_val),
    epochs=15,
    batch_size=64
)
```

## CNN with L2 Regularization

```python
def build_l2():
    model = models.Sequential([
        layers.Conv2D(32,(3,3),activation='relu',
                      kernel_regularizer=regularizers.l2(0.001),
                      input_shape=(32,32,3)),
        layers.MaxPooling2D(),
        layers.Conv2D(64,(3,3),activation='relu',
                      kernel_regularizer=regularizers.l2(0.001)),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(128,activation='relu',
                     kernel_regularizer=regularizers.l2(0.001)),
        layers.Dense(10,activation='softmax')
    ])
    return model

l2_model = build_l2()
l2_model.compile(optimizer='adam',
                 loss='sparse_categorical_crossentropy',
                 metrics=['accuracy'])

history_l2 = l2_model.fit(
    X_train_small,
    y_train_small,
    validation_data=(X_val, y_val),
    epochs=15,
    batch_size=64
)
```

## Data Augmentation using TensorFlow

```python
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip('horizontal'),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1)
])

plt.figure(figsize=(8,8))

for i in range(9):
    augmented = data_augmentation(X_train[:1])
    plt.subplot(3,3,i+1)
    plt.imshow(augmented[0])
    plt.axis('off')

plt.show()
```

## Model Comparison

```python
plt.figure(figsize=(10,6))

plt.plot(history_baseline.history['val_accuracy'], label='Baseline')
plt.plot(history_dropout.history['val_accuracy'], label='Dropout')
plt.plot(history_bn.history['val_accuracy'], label='BatchNorm')
plt.plot(history_l2.history['val_accuracy'], label='L2')

plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy')
plt.title('Validation Accuracy Comparison')
plt.legend()
plt.grid(True)
plt.show()
```

## Validation Accuracy Analysis

```python
results = {
    'Baseline': max(history_baseline.history['val_accuracy']),
    'Dropout': max(history_dropout.history['val_accuracy']),
    'BatchNorm': max(history_bn.history['val_accuracy']),
    'L2': max(history_l2.history['val_accuracy'])
}

for model, score in results.items():
    print(f'{model}: {score:.4f}')
```

## Best Overfitting Prevention Technique

```python
best_model = max(results, key=results.get)
print('Best model:', best_model)
```
